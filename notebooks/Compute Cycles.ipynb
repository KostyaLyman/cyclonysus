```python
import numpy as np
import dionysus
from sklearn import datasets
import matplotlib.pyplot as plt
from ripser import Rips

from cyclenysus import Cycler
```

```python
data, labels = datasets.make_circles(40, noise=0.1)
data = np.concatenate([data,np.random.random((30,2))+1])
```

```python
xs, ys = data[:,0], data[:,1]
plt.scatter(xs, ys)
plt.show()
```

# Construct filtration

```python
# Generate persistence diagram 
cycler = Cycler()
cycler.fit(data)

# Get top interval
longest = cycler.longest_intervals(1)[0]
cycle = cycler.get_cycle(longest)

vertices = cycler.order_vertices(cycle)
```

# Visualize raw cycle

```python
# Plot persistence diagram
rips = Rips()
rips.plot(cycler.barcode, label='$H_1$', show=False)
plt.title("Persistence Daigram for $H_1$")
```

```python
# Plot cycle
xs_v, ys_v = data[vertices][:,0], data[vertices][:,1]
plt.plot(xs_v, ys_v)

# Plot data
xs, ys = data[:,0], data[:,1]
plt.scatter(xs, ys)
plt.show()
```

# Show the top few cycles

```python
# Generate persistence diagram 
cycler = Cycler()
cycler.fit(data)

top_intervals = cycler.longest_intervals(3)
cycles = [cycler.get_cycle(interval) for interval in top_intervals]

# cycle = cycler.data_representation_of_cycle(best_cycle)
vertex_sets = [cycler.order_vertices(cycle) for cycle in cycles]
```

```python
# Plot cycle
for vertices in vertex_sets:
    xs_v, ys_v = data[vertices][:,0], data[vertices][:,1]
    plt.plot(xs_v, ys_v)

# Plot data
xs, ys = data[:,0], data[:,1]
plt.scatter(xs, ys)
plt.show()
```
